```python
import pandas as pd
import fakenewsutilities as fns
from sklearn.model_selection import train_test_split
import datetime
```

```python
df = pd.read_csv('tweets_labeled.csv')
df_clean = fns.wash_pandas_str(df)
df_clean.head()
```

```python
# get X and Y and split the data.
X = df_clean.iloc[:,0:2]
Y = df_clean.iloc[:,2]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 45, test_size  = 0.3) 
```

```python
#get the functions running time
starttime = datetime.datetime.now()
```

```python
train_df, fake_prob_prior = fns.naive_bayes_train(X_train, Y_train)
Word_frequency_feature = fns.naive_bayes_generate_feature(train_df, fake_prob_prior, X_test, Y_test)
```

```python
train_df_bigrm, fake_prob_prior = fns.naive_bayes_bigrm_train(X_train, Y_train)
bigram_frequency_feature = fns.naive_bayes_generate_feature_bigram(train_df_bigrm, fake_prob_prior, X_test, Y_test)
```

```python
endtime = datetime.datetime.now()
print (endtime - starttime)
```

The running time I listed in beginning was run on Visual Studio Code ipython.
Seems much faster than run on the notebook. 

```python
final_df = pd.DataFrame()
final_df['test_labels'] = Y_test
final_df['word_freq_2000'] = Word_frequency_feature
final_df['bigrm_freq_2000'] = bigram_frequency_feature
final_df.head()
```
